## **1.0. Preliminary Sizing of Columns Based on Axial Capacity**

---

### Conceptual basis and normative framework (ACI 318-19)

In this section, the **preliminary sizing of reinforced concrete columns based exclusively on axial capacity** is developed. The procedure assumes that the **factored axial demand $P_u$** acting on the column is known and provided by the user, typically obtained from a global structural analysis of the building.

At this stage, the objective is **not to perform a complete column design**, but rather to establish **reasonable initial cross-sectional dimensions** that are consistent with the expected axial demand and with the strength provisions of **ACI 318-19**. These preliminary dimensions will later be verified and refined through:
- axial–flexural interaction analysis,
- slenderness and second-order effects,
- confinement and detailing requirements.

All equations in this section are written in a form **compatible with the MKS-metric system (kgf–cm)**, as permitted by **ACI 318-19, Appendix C**, allowing consistent calculations using **kgf, cm, and kgf/cm²**, while reporting final results in **tons (T)** if desired.

> **Unit convention adopted in this section (consistent with Appendix C, ACI 318-19)**
> - Force for inputs/outputs: **tons-force (T)**  
> - Internal force unit for calculations: **kgf**  
> - Length for section geometry: **cm**  
> - Material stresses: **kgf/cm²**  
> - Conversions:  
>   - $1\,T = 1000\,kgf$  
>   - $1\,m = 100\,cm$

---

> **Important engineering note**  
> Preliminary sizing governed primarily by axial demand is appropriate **only when the structural system provides a significant lateral resistance through structural walls or a dual system (walls + frames)**. In such systems, columns are not the primary lateral load-resisting elements and axial forces tend to dominate their response.  
> This approach is **not applicable to pure moment-resisting frame systems**, where flexural demands control column dimensions and axial-based sizing would be unconservative.

---

#### **1.1. Nominal concentric axial strength $P_o$**

According to **ACI 318-19, Section 22.4.2.2**, the **nominal axial strength** of a reinforced concrete column subjected to **concentric compression** is defined as:

$$
P_o = 0.85\,f'_c\,(A_g - A_{st}) + f_y\,A_{st}
$$

where:
- $P_o$: nominal concentric axial strength $[kgf]$,  
- $f'_c$: concrete compressive strength $[kgf/cm^2]$,  
- $f_y$: yield strength of longitudinal reinforcement $[kgf/cm^2]$,  
- $A_g$: gross cross-sectional area of the column $[cm^2]$,  
- $A_{st}$: total area of longitudinal reinforcement $[cm^2]$.

This expression represents the **theoretical maximum axial capacity** under idealized concentric loading, combining the contributions of concrete and steel.

> **Engineering remark:**  
> Perfectly concentric loading does not occur in real structures due to unavoidable construction tolerances and load eccentricities. For this reason, the usable axial strength must be reduced, as addressed in subsequent sections.

---

#### **1.2. Nominal axial strength expressed in terms of reinforcement ratio $\rho$**

Defining the **longitudinal reinforcement ratio** as:

$$
\rho = \frac{A_{st}}{A_g}
$$

the nominal axial strength can be rewritten as:

$$
P_o = \left[\,0.85\,f'_c\,(1-\rho) + f_y\,\rho\,\right] A_g
$$

This formulation is particularly convenient for **preliminary sizing**, as it allows the designer to:
- assume a reasonable reinforcement ratio,
- directly estimate the required gross column area.

According to **ACI 318-19, Section 10.6.1.1**, the longitudinal reinforcement ratio must satisfy:

$$
0.01 \le \rho \le 0.08
$$

These limits ensure adequate constructability, confinement, and ductile behavior of reinforced concrete columns.

---

#### **1.3. Limitation of axial strength due to accidental eccentricity**

To account for unavoidable **accidental eccentricities**, **ACI 318-19, Section 22.4.2.1**, limits the usable axial strength of columns.

For **columns with transverse reinforcement (tied columns)**, the maximum nominal axial strength permitted for design is:

$$
P_{n,\max} = 0.80\,P_o
$$

This reduction represents a **practical upper bound** on axial resistance and implicitly accounts for eccentricities that cannot be avoided in construction and loading.

---

#### **1.4. Strength reduction factor $\phi$**

Design strength must further account for uncertainties in material properties, construction quality, and modeling assumptions through the **strength reduction factor $\phi$**.

For columns governed by **compression-controlled behavior**, the appropriate value is:

$$
\phi = 0.65
$$

according to **ACI 318-19, Table 21.2.2**, applicable to **tied columns**.

At the preliminary sizing stage, assuming compression-controlled behavior is both **appropriate and conservative**, particularly for columns dominated by axial load.

---

#### **1.5. Design condition and calculation of the gross concrete area $A_g$**

The fundamental design condition for axial-based preliminary sizing is:

$$
P_u \le \phi\,P_{n,\max}
$$

where:
- $P_u$ is the factored axial demand $[kgf]$.

Substituting the expressions introduced previously:

$$
P_u \le
\phi \cdot 0.80 \cdot
\left[\,0.85\,f'_c\,(1-\rho) + f_y\,\rho\,\right] A_g
$$

Solving for the required gross concrete area:

$$
A_g \ge
\frac{P_u}
{\phi \cdot 0.80 \cdot \left[\,0.85\,f'_c\,(1-\rho) + f_y\,\rho\,\right]}
$$

This expression provides a **direct and efficient estimate** of the minimum column area required to resist the axial demand at the preliminary stage.

> **Note on units:**  
> If $P_u$ is provided in tons, it must be converted to kgf before use:
> $$
> P_u\,[kgf] = P_u\,[T]\times 1000
> $$

---

#### **1.6. Preliminary cross-sectional dimension**

If a **square cross section** is adopted for simplicity and construction efficiency, the preliminary column dimension may be estimated as:

$$
b \approx \sqrt{A_g}
$$

where:
- $b$ is the side dimension of the square column $[cm]$.

Finally, the value of $b$ must be **rounded up to a practical construction dimension**, typically to multiples of **5 cm**, ensuring that the adopted section provides at least the required axial capacity.

At this stage, the column dimensions obtained should be understood strictly as **initial estimates**, which must later be verified against:
- axial–flexural interaction,
- slenderness effects,
- confinement and detailing requirements,

in subsequent stages of the structural design process.


#### **Calculations**

```python
Pu_T = 2970                                                                                                                         # Axial load in Tons (critical column load)
Vu_T = 138.93                                                                                                                         # Shear load in Tons (critical column shear)
Mu_T = 77.80                                                                                                                        # Bending moment in Ton-m (critical column moment)
fi = 0.65                                                                                                                           # Strength reduction factor for columns
fc = 500                                                                                                                            # Concrete compressive strength in kgf/cm^2
fy = 4200                                                                                                                           # Steel yield strength in kgd/cm^2
ro = 0.01                                                                                                                           # Flexural Reinforcement ratio (1% - 8% typical range)
rec = 4.00                                                                                                                          # Concrete cover in cm 

Pu = Pu_T * 1000                                                                                                                    # Axial load in Kgf (critical column load)
Vu = Vu_T * 1000                                                                                                                    # Shear load in Kgf (critical column shear)
Mu = Mu_T * 1000 * 100                                                                                                              # Bending moment in Kgf-cm (critical column moment)
```

* Cálculo de Ag

```python
Pu_df = pd.DataFrame({'Pu (Tons)': [Pu_T]})                                                                                         # DataFrame conversion for better visualization
Vu_df = pd.DataFrame({'Vu (Tons)': [Vu_T]})                                                                                         # DataFrame conversion for better visualization
Mu_df = pd.DataFrame({'Mu (Tons-m)': [Mu_T]})                                                                                       # DataFrame conversion for better visualization
                     
Ag = Pu / (fi * 0.80 * (0.85*fc*(1 - ro)) + fy * ro)                                                                                # Required gross area of the column cross-section in cm^2
Ag_df = pd.DataFrame({'Ag (cm2)': [Ag]})                                                                                            # DataFrame conversion for better visualization
print("\033[94m\033[1m"'=' * 120)
print("\033[94m\033[1m"
    "🏗️ Gross Area Required for Column Cross-Section")                                                                              
print('=' * 120)
Ag_df.head(len(Ag_df))                                                                                                              # Display the required gross area of the column cross-section                               
```

```python
b = np.sqrt(Ag)                                                                                                                     # Initial estimate for the width of the square column in cm       
B = np.ceil(b / 5.00) * 5.00                                                                                                        # Round up to the nearest 5 cm   
B_df = pd.DataFrame({'B (cm)': [B]})                                                                                                # DataFrame conversion for better visualization

fiPn = fi * ((0.85 * fc * (1 - ro)  + fy * ro) * B**2)                                                                              # Design axial capacity in Tons
utilP = Pu / fiPn                                                                                                                   # Axial load utilization ratio
fiPn_df = pd.DataFrame({'φPn (Tons)': [fiPn / 1000]})                                                                               # DataFrame conversion 
utilP_df = pd.DataFrame({'Utilization Pu/φPn': [utilP]})                                                                            # DataFrame conversion

print("\033[94m\033[1m"'=' * 120)                                                                                                   # Title
print("\033[94m\033[1m" "🏗️ Section")
print(f'Pu [T] = {Pu/1000:.2f}')
print(f'φPn [T]= {fiPn/1000:.2f}')
print(f'Util = {utilP:.2f}')
print("\033[94m\033[1m"'=' * 120)
B_df.head(len(B_df))                                                                                                                # Display the width of the square column cross-section
```

#### **1.7. Approximate Verification of Column Flexural Capacity**  
##### (Simplified estimation of $\phi M_n$ – ACI 318-19)

Once the **preliminary cross section of the column** has been defined based on axial capacity, it is necessary to verify that this section provides **sufficient flexural capacity to resist the factored bending moment demand $M_u$**, obtained from structural analysis.

At this stage, the objective is **not to perform a complete column design**, but to carry out a **screening-level verification** that allows clearly inadequate sections to be identified early in the design process.

The **rigorous code-based procedure** requires the construction of the **uniaxial axial–flexural interaction diagram $P$–$M$**, in accordance with **ACI 318-19, Section 22.4.3**.  
However, for **preliminary sizing purposes**, a **simplified and conservative estimation** of the flexural capacity is adopted, consisting of the evaluation of an approximate value of $\phi M_n$ under controlled and code-consistent assumptions.

All expressions in this section are written in a form **compatible with the MKS-metric system (kgf–cm)**, consistent with **ACI 318-19, Appendix C**.

---

##### **1.7.1. Normative background**

According to **ACI 318-19, Section 22.4.3**, the strength of columns subjected to bending must be evaluated considering the **interaction between axial load and bending moment**.

In preliminary sizing, this interaction is represented in an approximate manner by evaluating the flexural capacity under a **high axial compression state**, which is typical of columns in building structures, particularly when:
- structural walls are present, or
- the system behaves as a dual system (walls + frames).

Under these conditions, axial compression significantly influences the flexural response of the column.

---

##### **1.7.2. Assumptions adopted for the estimation of $M_n$**

To estimate the **nominal flexural strength $M_n$**, the following **ACI-consistent assumptions** are adopted:

- maximum concrete compressive strain:
  $$
  \varepsilon_{cu} = 0.003
  $$
  according to **ACI 318-19, Section 22.2.2.1**,
- plane sections remain plane after deformation,
- longitudinal reinforcement stress does not exceed the yield stress $f_y$,
- **uniaxial bending**,
- **rectangular or square cross section**,
- axial compression acting simultaneously with bending.

These assumptions correspond to the **strain compatibility approach** prescribed by the ACI and constitute the mechanical basis for flexural strength evaluation.

---

##### **1.7.3. Internal forces considered**

Under the assumptions stated above, the primary internal forces considered in the column cross section are:

- **Concrete compression force**:
  $$
  C_c = 0.85\,f'_c\,a\,b
  $$
  where:
  - $C_c$ is the resultant compressive force in concrete $[kgf]$,
  - $f'_c$ is the concrete compressive strength $[kgf/cm^2]$,
  - $a$ is the depth of the equivalent rectangular compression block $[cm]$,
  - $b$ is the width of the section $[cm]$,

- **Longitudinal steel force**:
  $$
  F_s = A_s\,f_s
  $$
  where:
  - $A_s$ is the area of longitudinal reinforcement $[cm^2]$,
  - $f_s$ is the steel stress, limited to $f_y$ when yielding occurs $[kgf/cm^2]$.

The formulation of the equivalent rectangular compression block follows **ACI 318-19, Section 22.2.2.4**.

---

##### **1.7.4. Approximate evaluation of the nominal flexural strength $M_n$**

The **nominal flexural strength** is obtained from equilibrium of internal moments:

$$
M_n = \sum F_i \, z_i
$$

where:
- $F_i$ are the internal forces $[kgf]$,
- $z_i$ are the corresponding lever arms $[cm]$.

For a column with **symmetrically distributed longitudinal reinforcement** and subjected to **uniaxial bending**, this expression may be written in simplified form as:

$$
M_n \approx C_c \cdot z_c + \sum F_{s,i} \cdot z_{s,i}
$$

where:
- $z_c$ is the lever arm between the concrete compression resultant and the reference axis,
- $z_{s,i}$ are the lever arms associated with the longitudinal reinforcement forces.

> **Engineering interpretation:**  
> This value represents **a single point on the axial–flexural interaction surface**, corresponding to a high axial compression level. It does **not** represent the full interaction behavior of the column.

The resulting $M_n$ is obtained in $[kgf\cdot cm]$ and may be converted to $[T\cdot m]$ as:

$$
M_n\,[T\cdot m] = \frac{M_n\,[kgf\cdot cm]}{100{,}000}
$$

---

##### **1.7.5. Design flexural capacity $\phi M_n$**

The **design flexural capacity** is obtained by applying the **strength reduction factor $\phi$**:

$$
\phi M_n
$$

For columns subjected to **high axial compression**, it is conservative and appropriate at the preliminary stage to adopt:

$$
\phi = 0.65
$$

according to **ACI 318-19, Table 21.2.2**, corresponding to **compression-controlled behavior in tied columns**.

---

##### **1.7.6. Verification condition for preliminary sizing**

The simplified flexural verification is expressed as:

$$
M_u \le \phi M_n
$$

where:
- $M_u$ is the factored bending moment demand $[kgf\cdot cm]$ (or $[T\cdot m]$ after conversion).

If this condition **is not satisfied**, the preliminary column section must be modified by:
- increasing the cross-sectional dimensions, and/or
- increasing the longitudinal reinforcement ratio,

while respecting the limits established in **ACI 318-19, Section 10.6.1.1**.

##### **Calculations**

```python
h = B                                                                                                                               # Column depth in cm (square section BxB)
b = B                                                                                                                               # Column width in cm

Ast = ro * Ag                                                                                                                       # Total longitudinal steel area in cm^2
As_t = 0.50 * Ast                                                                                                                   # Steel in tension side (cm^2) - assumed half
As_c = 0.50 * Ast                                                                                                                   # Steel in compression side (cm^2) - assumed half

d  = h - rec                                                                                                                        # Effective depth to tension steel in cm
dp = rec                                                                                                                            # Depth to compression steel in cm

Pn = fiPn / fi                                                                                                                      # Nominal axial force in Kgf (so that fi*Pn ≈ Pu)

a = Pn / (0.85 * fc * b)                                                                                                            # Depth of equivalent stress block in cm
a = np.minimum(a, 0.99 * h)                                                                                                         # Cap 'a' to remain within the section

Cc = 0.85 * fc * b * a                                                                                                              # Concrete compression resultant in Kgf
Ts = As_t * fy                                                                                                                      # Tension steel force in Kgf
Cs = As_c * fy                                                                                                                      # Compression steel force in Kgf

Mn = (Cc * (h/2 - a/2)) + (Ts * (d  - h/2)) + (Cs * (h/2 - dp))                                                                     # Nominal moment in Kgf-cm
fiMn = fi * Mn                                                                                                                      # Design moment capacity in Kgf-cm

fiMn_df = pd.DataFrame({'φMn (Ton-m)': [fiMn / (1000*100)],'Mu (Ton-m)': [Mu_T],'Utilization Mu/φMn': [Mu / fiMn]})                 # DataFrame conversion for better visualization

print("\033[94m\033[1m"
      "🏗️ Flexural Capacity Check (Simplified)\n"
      "------------------------------------------------------------\n"
      f"φMn = {fiMn / (1000*100):.3f} Ton-m\n"
      f"Mu   = {Mu_T:.3f} Ton-m\n"
      f"Mu/φMn = {Mu/fiMn:.3f}  ->  {'OK ✅' if Mu <= fiMn else 'NOT OK ❌'}\n"
      "\033[0m")

fiMn_df.head(len(fiMn_df))                                                                                                          # Display simplified flexural verification results
```

## **2.0. Preliminary Sizing of Reinforced Concrete Beams**

---

### Conceptual basis and design inputs

In this section, the **preliminary sizing of reinforced concrete beams** is developed assuming that the **factored bending moment $M_u$** and the **factored shear force $V_u$** are provided by the user, typically obtained from a global structural analysis.

All equations are written in a form **compatible with the MKS-metric system (kgf–cm)** as permitted by **ACI 318-19, Appendix C**, so that the calculations can be carried out consistently using **kgf, cm, and kgf/cm²**, while still reporting final results in **T** and **T·m** when desired.

> **Unit convention adopted in this section of the notebook (consistent with Appendix C, ACI 318-19)**
> - Force for inputs/outputs: **tons-force (T)**  
> - Internal force unit for calculations: **kgf**  
> - Length for section geometry: **cm**  
> - Material stresses: **kgf/cm²**  
> - Conversions:  
>   - $1\,T = 1000\,kgf$  
>   - $1\,m = 100\,cm$  
>   - $1\,T\cdot m = 1000\times 100 = 100{,}000\;kgf\cdot cm$

---

#### **2.1. Design philosophy for preliminary beam sizing**

Beams in building structures are governed primarily by **flexure and shear**, with axial forces generally negligible. Accordingly, preliminary beam sizing follows this sequence:

1. Select an initial beam geometry.  
2. Size the section to resist the **factored bending moment $M_u$**.  
3. Verify compliance with **minimum flexural reinforcement requirements**.  
4. Check compatibility with the **factored shear force $V_u$**.

---

#### **2.2. Initial estimate of beam geometry**

Since **ACI 318-19 does not prescribe geometric proportions**, the initial beam depth is selected based on engineering practice:

- **Simply supported beams**:
$$
h \approx \frac{L}{10} \text{ to } \frac{L}{12}
$$

- **Continuous beams**:
$$
h \approx \frac{L}{12} \text{ to } \frac{L}{15}
$$

where:
- $h$: total beam depth $[cm]$  
- $L$: clear span $[cm]$

Beam width is preliminarily selected as:

$$
b \approx (0.3 \text{ to } 0.5)\,h
$$

where $b$ is in $[cm]$.

---

#### **2.3. Effective depth**

The effective depth is approximated as:

$$
d \approx h - r
$$

where:
- $d$: effective depth $[cm]$  
- $r$: cover + stirrup + half bar diameter $[cm]$

---

#### **2.4. Flexural strength requirement**

According to **ACI 318-19, Chapter 21**, the flexural design condition is:

$$
M_u \le \phi M_n
$$

For tension-controlled nonprestressed beams:

$$
\phi = 0.90
$$

(ACI 318-19, Table 21.2.2)

> **Note on units:** In the calculations, $M_n$ is first obtained in $kgf\cdot cm$ and then converted to $T\cdot m$ by:
> $$
> M_n\,[T\cdot m] = \frac{M_n\,[kgf\cdot cm]}{100{,}000}
> $$

---

#### **2.5. Nominal flexural strength (MKS-metric form)**

Using the equivalent rectangular stress block:

**Depth of compression block**  
(ACI 318-19, Section 22.2.2.4)

$$
a = \frac{A_s f_y}{0.85 f'_c b}
$$

where:
- $a$: depth of equivalent compression block $[cm]$  
- $A_s$: tensile reinforcement area $[cm^2]$  
- $f_y$: steel yield stress $[kgf/cm^2]$  
- $f'_c$: concrete compressive strength $[kgf/cm^2]$  
- $b$: beam width $[cm]$  

**Nominal moment capacity**

$$
M_n = A_s f_y \left(d - \frac{a}{2}\right)
$$

where:
- $M_n$ is obtained in $[kgf\cdot cm]$ (and may be converted to $[T\cdot m]$ as noted above).

---

#### **2.6. Reinforcement ratio formulation**

Define the tensile reinforcement ratio:

$$
\rho = \frac{A_s}{b d}
$$

where:
- $\rho$ is dimensionless,
- $A_s$ in $[cm^2]$, $b$ and $d$ in $[cm]$.

Substituting $A_s = \rho b d$ into the flexural model yields:

$$
M_n = \rho b f_y d^2
\left(1 - \frac{\rho f_y}{1.70 f'_c}\right)
$$

Thus, the preliminary flexural sizing condition becomes:

$$
M_u \le
\phi \rho b f_y d^2
\left(1 - \frac{\rho f_y}{1.70 f'_c}\right)
$$

Solving for required effective depth:

$$
d \ge
\sqrt{
\frac{M_u}
{\phi \rho b f_y
\left(1 - \frac{\rho f_y}{1.70 f'_c}\right)}
}
$$

> **Important:** For this step, $M_u$ must be used in $kgf\cdot cm$:
> $$
> M_u\,[kgf\cdot cm] = M_u\,[T\cdot m]\times 100{,}000
> $$

and:

$$
h \approx d + r
$$

---

#### **2.7. Minimum flexural reinforcement (ACI–MKS form)**

According to **ACI 318-19, Section 9.6.1.2** and the unit conversions provided in **ACI 318-19, Appendix C**, the minimum tensile reinforcement for beams in **MKS-metric units** is:

$$
A_{s,\min} =
\max
\left(
\frac{0.80\sqrt{f'_c}}{f_y} b d,
\;
\frac{14}{f_y} b d
\right)
$$

where:
- $f'_c$ and $f_y$ are in $[kgf/cm^2]$,
- $b$ and $d$ are in $[cm]$,
- $A_{s,\min}$ is obtained in $[cm^2]$.

Requirement:

$$
A_s \ge A_{s,\min}
$$

This applies to:
- bottom reinforcement in positive moment regions,
- top reinforcement in negative moment regions.

#### **Calculations**

```python
Nu_T = 0.00                                                                                                                         # Axial load in Tons (critical beam load)
Vu_T = 90.0                                                                                                                        # Shear load in Tons (critical beam shear)
Mu_T = 120.0                                                                                                                       # Bending moment in Ton-m (critical beam moment)
fiF = 0.90                                                                                                                          # Strength reduction factor for beams (Flexure)
fiC = 0.75                                                                                                                          # Strength reduction factor for beams (Shear)
lamda = 1.00                                                                                                                        # Modification factor for shear
rec = 4.00                                                                                                                          # Concrete cover in cm
L = 1000.00                                                                                                                         # Beam span in cm
db = 1.20                                                                                                                           # Diameter of shear reinforcement in cm

Vu = Vu_T * 1000                                                                                                                    # Shear load in Kgf (critical beam shear)
Mu = Mu_T * 1000 * 100                                                                                                              # Bending moment in Kgf-cm (critical beam moment)
Nu = Nu_T * 1000                                                                                                                    # Axial load in Kgf (critical beam load)
```

```python
h = L / 12                                                                                                                          # Initial estimate for beam depth in cm (L/12 rule of thumb)
hinicial_df = pd.DataFrame({'h initial (cm)': [h]})                                                                                 # DataFrame conversion for better visualization
b = 0.50 * h                                                                                                                        # Initial estimate for beam width in cm (b = 0.5*h)
b = np.ceil(b / 5.00) * 5.00                                                                                                        # Round up to nearest 5 cm
binicial_df = pd.DataFrame({'b initial (cm)': [b]})                                                                                 # DataFrame conversion for better visualization

Asmin = np.maximum(0.80 * np.sqrt(fc) / fy * b * h, 14 / fy * b * h)                                                                # Minimum steel area in cm^2

ro = 2 * Asmin / (b * h)                                                                                                            # Flexural Reinforcement ratio based on minimum steel area
d = np.sqrt(Mu / (fiF * ro * b * fy * (1 - (ro * fy) / (1.7 - fc))))                                                                # Effective depth in cm 
print(d)
h = d + rec                                                                                                                         # Adjusted beam depth in cm  
h = np.ceil(h / 5.00) * 5.00                                                                                                        # Round up to nearest 5 cm
d = h - rec                                                                                                                         # Effective depth to tension steel in cm
b = 0.50 * h                                                                                                                        # Adjusted beam width in cm (b = 0.5*h)
b = np.ceil(b / 5.00) * 5.00                                                                                                        # Round up to nearest 5 cm

ro = 1.30 * 2 * Asmin / (b * h)                                                                                                     # Recalculate flexural Reinforcement ratio based on adjusted depth

h_df = pd.DataFrame({'h (cm)': [h]})                                                                                                # DataFrame conversion for better visualization
b_df = pd.DataFrame({'b (cm)': [b]})                                                                                                # DataFrame conversion for better visualization   
d_df = pd.DataFrame({'d (cm)': [d]})                                                                                                # DataFrame conversion for better visualization  
Asmin_df = pd.DataFrame({'Asmin (cm2)': [2 * Asmin]})                                                                               # DataFrame conversion for better visualization
ro_T_df = pd.DataFrame({'Ratio ro (%) [Tension]': [ro / 2 * 100]})                                                                  # DataFrame conversion for better visualization
ro_C_df = pd.DataFrame({'Ratio ro (%) [Compression]': [ro / 2 *100]})                                                               # DataFrame conversion for better visualization    

Geometry_df = pd.concat([hinicial_df, binicial_df, h_df, b_df, d_df, Asmin_df, ro_T_df, ro_C_df], axis = 1)                                                   # Combine geometry results into a single DataFrame  
print("\033[94m\033[1m"
    "🏗️ Beam Geometry \n"
    "----------------------------------------------\n")
Geometry_df.head(len(Geometry_df))                                                                                                  # Display beam geometry results                                                                                                     
             
```

#### **2.8. Shear strength requirement**

The shear design condition is:

$$
V_u \le \phi V_n
$$

with:

$$
\phi = 0.75
$$

(ACI 318-19, Table 21.2.2)

and:

$$
V_n = V_c + V_s
$$

(ACI 318-19, Section 22.5.1)

> **Note on units:** In the calculations, $V_u$, $V_c$, and $V_s$ are handled in $[kgf]$:
> $$
> V_u\,[kgf] = V_u\,[T]\times 1000
> $$

---

#### **2.9. Concrete shear strength $V_c$ (ACI 318-19, 22.5.5.1(a) – MKS-metric form)**

For nonprestressed beams, **ACI 318-19, Section 22.5.5.1(a)** in **MKS-metric form (Appendix C)** is:

$$
V_c =
\left(
0.53\,\lambda\sqrt{f'_c}
+
\frac{N_u}{6A_g}
\right)\, b_w\, d
$$

where:
- $V_c$: concrete shear strength $[kgf]$  
- $f'_c$: concrete strength $[kgf/cm^2]$  
- $b_w$: web width (for rectangular beams, $b_w=b$) $[cm]$  
- $d$: effective depth $[cm]$  
- $N_u$: factored axial force on the member $[kgf]$ (compression taken as positive)  
- $A_g$: gross area of section $[cm^2]$  

**Lightweight concrete factor $\lambda$**  
According to **ACI 318-19, Section 19.2.4**, $\lambda$ is the modification factor used to reduce strength expressions when lightweight concrete is used. For normalweight concrete:

$$
\lambda = 1.0
$$

For lightweight concrete, $\lambda$ must be taken as specified by the code (depending on the concrete type and density), and should be defined explicitly by the user if applicable.

Converted to tons (if desired):

$$
V_c\,[T] = \frac{V_c\,[kgf]}{1000}
$$

---

#### **2.10. Required stirrup contribution**

If:

$$
V_u > \phi V_c
$$

then required stirrup strength is:

$$
V_s \ge \frac{V_u}{\phi} - V_c
$$

For vertical stirrups (ACI 318-19, Section 22.5.10.5):

$$
V_s = \frac{A_v f_y d}{s}
$$

leading to:

$$
\frac{A_v}{s} \ge
\frac{\frac{V_u}{\phi} - V_c}{f_y d}
$$

where:
- $A_v$: area of shear reinforcement within spacing $s$ $[cm^2]$  
- $s$: stirrup spacing $[cm]$  
- $f_y$: transverse reinforcement yield stress $[kgf/cm^2]$  
- $d$: effective depth $[cm]$


#### **2.12. Maximum Spacing of Transverse Reinforcement (Stirrups) in Beams**

Once the required transverse reinforcement has been defined, the **spacing of stirrups must comply with the maximum spacing limits prescribed by ACI 318-19**, independently of the shear strength calculations.

These limits are intended to ensure:
- effective interception of diagonal cracks,
- proper development of the truss mechanism,
- confinement of longitudinal reinforcement,
- and adequate crack control.


According to **ACI 318-19, Section 22.5.10.4.2**, for **nonprestressed reinforced concrete beams with vertical stirrups**, the **center-to-center spacing $s$ of transverse reinforcement** shall not exceed the smallest of the following limits:

$$
s \le \min
\left(
\frac{d}{2},
\; 6\,d_b,
\; 60\;\text{cm}
\right)
$$

where:
- $s$ = stirrup spacing $[cm]$,  
- $d$ = effective depth of the beam $[cm]$,  
- $d_b$ = diameter of the **smallest longitudinal reinforcement bar** enclosed by the stirrup $[cm]$.

This requirement explicitly recognizes that **crack control and confinement effectiveness depend not only on the beam depth, but also on the size of the longitudinal reinforcement**.

```python
db = 2.5                                                                                                                           # Diameter of Flexure reinforcement in cm
dbC = 1.20                                                                                                                          # Diameter of Shear reinforcement in cm
S = np.minimum( np.minimum( d / 2 , 6 * db), 60)                                                                                    # Stirrup spacing in cm (max 60 cm, min 6*db, max d/2)
S = 5 * np.floor(S / 5)                                                                                                             # Round down to nearest 5 cm
Vc = (0.53 * lamda * np.sqrt(fc) + Nu / (6 * b * h)) * b * d                                                                        # Shear capacity in Kgf
Av = (Vu / fiC - Vc) / (fy * d) * S                                                                                                 # Required shear reinforcement area in cm^2
num_legs_stirrups = np.maximum(2, np.ceil(Av / (np.pi * dbC**2 / 4)))                                                                # Number of stirrup legs (at least 2) 
Av = num_legs_stirrups * (np.pi * dbC**2 / 4)                                                                                        # Adjusted shear reinforcement area in cm^2                                                                                 

Vs = Av * fy * d / S                                                                                                                # Shear strength provided by stirrups in Kgf
fiVs = fiC * Vs                                                                                                                     # Design shear strength provided by stirrups in Kgf
fiVc = fiC * Vc                                                                                                                     # Design shear strength provided by concrete in Kgf
fiVn = (fiVc + fiVs)                                                                                                                # Design shear capacity in Kgf

Av_df = pd.DataFrame({'Av (cm2)': [Av]})                                                                                            # DataFrame conversion for better visualization
S_df = pd.DataFrame({'Stirrup Spacing S (cm)': [S]})                                                                                # DataFrame conversion for better visualization
stirrups_df = pd.DataFrame({'Number of Stirrup Legs': [num_legs_stirrups]})                                                         # DataFrame conversion for better visualization
fiVc_df = pd.DataFrame({'φVc (Ton)': [fiVc / 1000]})                                                                                # DataFrame conversion for better visualization
fiVs_df = pd.DataFrame({'φVs (Ton)': [fiVs / 1000]})                                                                                # DataFrame conversion for better visualization
fiVn_df = pd.DataFrame({'φVn (Ton)': [fiVn / 1000]})                                                                                # DataFrame conversion for better visualization
Nu_T_df = pd.DataFrame({'Nu (Tons)': [Nu_T]})                                                                                       # DataFrame conversion for better visualization
Vu_T_df = pd.DataFrame({'Vu (Tons)': [Vu_T]})                                                                                       # DataFrame conversion for better visualization
Mu_T_df = pd.DataFrame({'Mu (Tons-m)': [Mu_T]})                                                                                     # DataFrame conversion for better visualization

print("\033[94m\033[1m"
      "🏗️ Shear Capacity Check (Simplified)\n"
      "------------------------------------------------------------\n"
      f"φVn = {fiVn / (1000):.3f} Ton\n"
      f"Vu   = {Vu_T:.3f} Ton\n"
      f"Vu/φVn = {Vu/fiVn:.3f}  ->  {'OK ✅' if Vu <= fiVn else 'NOT OK ❌'}\n"
      "\033[0m")
```

## **3.0. Preliminary Sizing of Two-Way Reinforced Concrete Slabs Supported on Drop Beams**

---

### Conceptual basis and design inputs (ACI 318-19 framework)

This section develops the **preliminary sizing of two-way reinforced concrete slabs supported on beams (drop beams)**. In this structural system, the slab behaves as a **two-way plate spanning between beams**, and the beams provide the primary line supports along the panel edges.

At this stage, the slab is first idealized as a **solid slab panel**, because this provides a clear baseline in terms of stiffness and strength. Once an acceptable solid-slab thickness is established, the section then shows the **explicit mathematical procedure** to transition to a **waffle (ribbed) slab** while maintaining a comparable flexural stiffness level.

The following are assumed to be provided by the user:
- clear spans between beams in both directions: $l_{n,x}$ and $l_{n,y}$,
- reinforcement yield strength $f_y$,
- concrete strength $f'_c$ (mainly for later flexural checks),
- preliminary factored moments per unit width from global analysis (if available): $M_{u,x}$ and $M_{u,y}$.

All expressions are consistent with the **MKS-metric (kgf–cm)** convention used in this notebook, while the ACI slab-thickness expressions are presented in their standard ACI form (dimensionally consistent) with explicit guidance on unit use.

> **Unit convention adopted in this section of the notebook**
> - Force: **T** (reported), **kgf** (internal)
> - Length for slab geometry: **cm**
> - Stresses: **kgf/cm²**
> - Conversions:
>   - $1\,T = 1000\,kgf$
>   - $1\,m = 100\,cm$
>   - $1\,T\cdot m = 100{,}000\;kgf\cdot cm$

---

#### **3.1. Two-way slab condition**

A slab panel supported on beams is treated as **two-way** when:

$$
\beta = \frac{l_{n,long}}{l_{n,short}} \le 2.0
$$

where:
- $l_{n,long}$ = longer clear span between beams [cm]  
- $l_{n,short}$ = shorter clear span between beams [cm]  
- $\beta$ = aspect ratio (dimensionless)

---

#### **3.2. Clear span definition**

For slab minimum thickness checks, the clear span is taken as:

$$
l_n = \text{clear distance between faces of supports (beams)}
$$

and the governing span for thickness selection is typically:

$$
l_{n,short} = \min(l_{n,x},\,l_{n,y})
$$

---

#### **3.3. Minimum slab thickness for two-way slabs with beams (ACI 318-19, Table 8.3.1.2)**

For **nonprestressed two-way slabs with beams spanning between supports on all sides**, **ACI 318-19, Table 8.3.1.2** permits selection of slab thickness **without explicit deflection calculations**, provided the table limits are satisfied. 

The ACI expression is written in terms of:
- $l_n$ = clear span in the direction considered
- $\beta$ = long-to-short clear-span ratio
- $\alpha_m$ = average beam-to-slab stiffness ratio on all sides of the panel (dimensionless)

The **general ACI form** used for slabs with beams is:

$$
h \ge \frac{l_n\left(0.8 + \frac{f_y}{1400}\right)}{D(\beta,\alpha_m)}
$$

where:
- $h$ = slab thickness [same length units as $l_n$]
- $f_y$ is used in **MPa** in this ACI expression (ACI standard SI form)
- $D(\beta,\alpha_m)$ is a denominator that depends on the stiffness level of beams, as defined by the table.

> **How to use $f_y$ consistently in this notebook**  
> Because this notebook uses stresses in **kgf/cm²**, convert:
> $$
> f_y[\text{MPa}] = \frac{f_y[\text{kgf/cm}^2]}{10.197}
> $$
> and then use $f_y[\text{MPa}]$ in $\left(0.8+\frac{f_y}{1400}\right)$.

---

##### **3.3.1. Denominator according to beam stiffness level**

ACI classifies the expression depending on $\alpha_m$ (panel beam-to-slab stiffness level). In particular, one of the common cases (used when the panel has sufficiently stiff beams) is:

$$
D(\beta,\alpha_m) = 36 + 9\beta
$$

leading to:

$$
h \ge \frac{l_n\left(0.8 + \frac{f_y}{1400}\right)}{36 + 9\beta}
$$

This form appears directly in worked ACI-based slab-with-beams examples and is consistent with **ACI 318-19, Table 8.3.1.2** usage.

> **Practical note:** in preliminary sizing, when the beam system is clearly stiff (deep drop beams), this expression provides a rational starting thickness. In later stages, the correct $\alpha_m$ category must be confirmed for the panel.

---

#### **3.4. Adopted preliminary solid-slab thickness**

Once the minimum thickness is computed, adopt:

$$
h_{solid} = \text{rounded-up practical thickness (e.g., to 1 cm or 5 mm)}
$$

and define:

$$
d_{solid} \approx h_{solid} - r
$$

where $r$ accounts for cover and half bar diameter (preliminary value acceptable).


#### **3.5. Preliminary flexural reinforcement concept (two orthogonal design strips)**

For preliminary flexural sizing, the slab is treated using **1 m design strips** in both directions:

$$
b = 100\;cm
$$

with user-defined factored moments per unit width:

$$
M_{u,x}\,[T\cdot m/m],\qquad M_{u,y}\,[T\cdot m/m]
$$

Each strip must satisfy:

$$
M_u \le \phi M_n,\qquad \phi = 0.90
$$

(ACI 318-19, Table 21.2.2)

Moments are converted to internal units as:

$$
M_u[kgf\cdot cm] = M_u[T\cdot m]\times 100{,}000
$$

---

#### **3.6. Baseline stiffness of the solid slab (reference for waffle transition)**

To transition from solid slab to waffle slab in a rational way, first compute the **reference flexural stiffness per unit width** of the solid slab (same concrete, so $E_c$ cancels in comparisons).

For a 1 m strip ($b=100$ cm), the gross moment of inertia is:

$$
I_{solid} = \frac{b\,h_{solid}^3}{12}
= \frac{100\,h_{solid}^3}{12}
\qquad [cm^4]
$$

This $I_{solid}$ will be used as the **stiffness benchmark**.

---

#### **3.7. Transition to waffle slab (explicit mathematics based on stiffness equivalence)**

A waffle slab (ribbed slab) may be idealized, in each direction, as a set of **parallel ribs** (T-beams) spaced at $s_r$, connected by a topping slab of thickness $h_f$.

Define waffle geometry:
- $s_r$ = rib spacing [cm]
- $b_w$ = rib (web) width [cm]
- $h_f$ = topping thickness [cm]
- $h_{tot}$ = total waffle thickness [cm]
- $h_r = h_{tot} - h_f$ = rib depth below topping [cm]

---

##### **3.7.1. Equivalent inertia per 1 m width of waffle slab**

Compute the gross inertia of **one rib T-section** about its centroid:

1) Areas:
$$
A_f = b_f h_f,\qquad A_w = b_w h_r
$$

where for preliminary stiffness:
- take $b_f \approx s_r$ (tributary flange width per rib) for a 1D stiffness idealization.

2) Centroid location from the top fiber:
$$
\bar{y} =
\frac{A_f\left(\frac{h_f}{2}\right) + A_w\left(h_f+\frac{h_r}{2}\right)}
{A_f + A_w}
$$

3) Inertia of components about their own centroids:
$$
I_f = \frac{b_f h_f^3}{12},\qquad
I_w = \frac{b_w h_r^3}{12}
$$

4) Parallel-axis theorem to the T-section centroid:
$$
I_{T} = I_f + A_f\left(\bar{y}-\frac{h_f}{2}\right)^2
      + I_w + A_w\left(h_f+\frac{h_r}{2}-\bar{y}\right)^2
\qquad [cm^4]
$$

Now convert to an **equivalent inertia per 1 m width** (100 cm) by recognizing that in 100 cm there are $\frac{100}{s_r}$ ribs:

$$
I_{waffle,eq} = I_T \left(\frac{100}{s_r}\right)
\qquad [cm^4 \; \text{per 1 m strip}]
$$

---

##### **3.7.2. Stiffness-based equivalence criterion (solid → waffle)**

To ensure that the waffle slab is not significantly more flexible than the baseline solid slab, impose:

$$
I_{waffle,eq} \ge \eta\, I_{solid}
$$

where:
- $\eta \approx 1.0$ is a strict stiffness match (recommended for a first transition),
- $\eta \approx 0.9$ may be used if a small stiffness reduction is acceptable and will be checked later by deflection analysis.

Substituting:

$$
I_T \left(\frac{100}{s_r}\right) \ge \eta\left(\frac{100\,h_{solid}^3}{12}\right)
$$

This inequality provides a **direct mathematical rule** to select $h_f$, $s_r$, $b_w$, and $h_{tot}$ (through $h_r$) so that the waffle slab maintains comparable stiffness to the solid slab baseline.

> **Engineering note:** This is the key step that makes the transition objective and defensible: the waffle option must “pay back” the reduction in solid concrete by providing equivalent structural depth and rib inertia.

---

#### **3.8. Calculation of the Average Beam-to-Slab Stiffness Ratio $\alpha_m$ (ACI 318-19, Table 8.3.1.2)**

Once the slab system has been preliminarily defined and the **waffle slab geometry is known** (i.e., $h_f$, $h_{tot}$, $s_r$, and rib dimensions), it is possible to compute the **average beam-to-slab stiffness ratio $\alpha_m$**, which is required to correctly select the appropriate expression in **ACI 318-19, Table 8.3.1.2** for minimum slab thickness.

At the preliminary stage, $\alpha_m$ is used to quantify how stiff the supporting beams are relative to the slab, and therefore how effective the beams are in controlling slab deflections.

---

##### **3.8.1. Definition of $\alpha_m$**

According to **ACI 318-19 (Table 8.3.1.2 context)**, the stiffness parameter for a two-way slab panel supported on beams is defined as an average value:

$$
\alpha_m = \frac{\alpha_x + \alpha_y}{2}
$$

where:
- $\alpha_x$ is the beam-to-slab stiffness ratio for the pair of beams parallel to the $x$ direction (i.e., beams supporting slab bending in the $x$ span),
- $\alpha_y$ is the beam-to-slab stiffness ratio for the pair of beams parallel to the $y$ direction.

---

##### **3.8.2. Beam-to-slab stiffness ratio in each direction**

For each direction, the stiffness ratio is computed as:

$$
\alpha = \frac{E_c I_b}{E_c I_s}
= \frac{I_b}{I_s}
$$

where:
- $I_b$ is the moment of inertia of the supporting beam (about its strong axis),
- $I_s$ is the effective slab moment of inertia for the slab strip tributary to that beam.

Since the beam and slab are made of the same concrete, the elastic modulus $E_c$ cancels out in preliminary stiffness comparisons.

Thus, for preliminary calculations:

$$
\alpha = \frac{I_b}{I_s}
$$

---

##### **3.8.3. Slab stiffness $I_s$ for a waffle slab panel**

Because the slab has been defined as a **waffle slab**, the appropriate slab inertia to represent the stiffness of the slab strip is the **equivalent inertia per unit width**, previously computed in Section **3.7**:

$$
I_{s} \equiv I_{waffle,eq}
$$

where:
- $I_{waffle,eq}$ is the equivalent inertia of the waffle slab per 1 m strip $[cm^4]$.


---

##### **3.8.4. Beam stiffness $I_b$**

The beam gross moment of inertia about its centroidal axis may be approximated as:

$$
I_b = \frac{b_b h_b^3}{12}
$$

where:
- $b_b$ is the beam width $[cm]$,
- $h_b$ is the beam total depth $[cm]$.

At the preliminary stage, it is acceptable to use the gross inertia of the beam. In detailed design, cracked section effects may be considered depending on the level of refinement required.

---

##### **3.8.5. Computing $\alpha_x$, $\alpha_y$, and $\alpha_m$**

For each direction:

$$
\alpha_x = \frac{I_{b,x}}{I_{s}}
\qquad\qquad
\alpha_y = \frac{I_{b,y}}{I_{s}}
$$

and finally:

$$
\alpha_m = \frac{\alpha_x + \alpha_y}{2}
$$


#### **Calculations**

```python
Mu_T = 82.157                                                                                                                       # Bending moment in Ton-m (critical slab moment)
fiF = 0.90                                                                                                                          # Strength reduction factor for slabs (Flexure)
rec = 2.50                                                                                                                          # Concrete cover in cm
L_long = 750.00                                                                                                                     # Longer span of slab in cm
L_short = 700.00                                                                                                                    # Shorter span of slab in cm
bw = 100.00                                                                                                                         # Width of slab strip in cm
sr = 50.00                                                                                                                          # Distance between slab strips in cm
bf = sr                                                                                                                             # Width of slab strip in cm
bst = 10.00                                                                                                                         # Width of slab strip in cm
hf = 5.00                                                                                                                           # Thickness of slab flange in cm
htot = 30.00                                                                                                                        # Tentative total slab thickness in cm
n = 0.90                                                                                                                            # Modification factor for slabs   

fy_MPA = fy / 10.1972                                                                                                               # Steel yield strength in MPa                                                                                                                                                                                               
Mu = Mu_T * 1000 * 100                                                                                                              # Bending moment in Kgf-cm (critical slab moment)
```

```python
Beta = np.minimum(L_short / L_long, 2.00)                                                                                           # Aspect ratio of slab spans
Ln = np.maximum(L_long, L_short)                                                                                                    # Longer span of slab in cm

hsolid = Ln * (0.80 + fy_MPA / 1400) / (36 + 9 * Beta)                                                                              # Initial estimate for slab thickness in cm
hsolid = np.ceil(hsolid / 5.00) * 5.00                                                                                              # Round up to nearest 5 cm

h_df = pd.DataFrame({'h (cm)': [hsolid]})                                                                                           # DataFrame conversion for better visualization


print("\033[94m\033[1m"
    "🏗️ Slab thickness \n"
    "----------------------------------------------\n")
h_df.head(len(h_df))                                                                                                                # Display initial estimate for slab thickness
```

```python
Isolid = (bw * hsolid**3) / 12                                                                                                      # Moment of inertia of solid slab strip in cm^4

hr = htot - hf                                                                                                                      # Rib depth below topping [cm]  
Af = bf * hf                                                                                                                        # Area of slab flange in cm^2  
Aw = bst * hr                                                                                                                        # Area of slab web in cm^2
AT = Af + Aw                                                                                                                        # Total area of T-slab in cm^2

y = (Af * (hf /2) + Aw * (hf + hr / 2)) / (Af + Aw)                                                                                 # Neutral axis depth from bottom of slab in cm  

If = (bf * hf**3) / 12                                                                                                              # Moment of inertia of slab flange in cm^4
Iw = (bst * hr**3) / 12                                                                                                              # Moment of inertia of slab web in cm^4

IT = If + Af * (y - hf / 2)**2 + Iw + Aw * (hf + hr / 2 - y)**2                                                                     # Total moment of inertia of T-slab in cm^4

Iwaffle_eq = IT * (100 / sr)
Slab_ratio = Iwaffle_eq / Isolid                                                                                                    # Ratio of waffle slab to solid slab moment of inertia


Is = IT                                                                                                                  # Equivalent moment of inertia of waffle slab strip in cm^4 in bwidth
Ibx = b * h**3 / 12                                                                                                                 # Moment of inertia of the beam in cm^4 (x direction)
Iby = Ibx                                                                                                                           # Moment of inertia of the beam in cm^4 (y direction)
alfax = Ibx / Is                                                                                                                    # Stiffness ratio in x direction
alfay = Iby / Is                                                                                                                    # Stiffness ratio in y direction

alfam = (alfax + alfay) / 2                                                                                                         # Average stiffness ratio

bw_df = pd.DataFrame({'bw (cm)': [bw]})                                                                                             # DataFrame conversion for better visualization   
sr_fd = pd.DataFrame({'sr (cm)': [sr]})                                                                                             # DataFrame conversion for better visualization
bf_df = pd.DataFrame({'bf (cm)': [bf]})                                                                                             # DataFrame conversion for better visualization
bst_df = pd.DataFrame({'bst (cm)': [bst]})                                                                                          # DataFrame conversion for better visualization
hf_df = pd.DataFrame({'hf (cm)': [hf]})                                                                                             # DataFrame conversion for better visualization
htot_df = pd.DataFrame({'htot (cm)': [htot]})                                                                                       # DataFrame conversion for better visualization
hr_df = pd.DataFrame({'hr (cm)': [hr]})                                                                                             # DataFrame conversion for better visualization
AT_df = pd.DataFrame({'AT (cm2)': [AT]})                                                                                            # DataFrame conversion for better visualization
Iwaffle_eq_df = pd.DataFrame({'Iwaffle_eq (cm4)': [Iwaffle_eq]})                                                                    # DataFrame conversion for better visualization
Isolid_df = pd.DataFrame({'Isolid (cm4)': [Isolid]})                                                                                # DataFrame conversion for better visualization
Slab_ratio_df = pd.DataFrame({'Slab_ratio': [Slab_ratio]})                                                                          # DataFrame conversion for better visualization
alfam_df = pd.DataFrame({'alfam': [alfam]})                                                                                         # DataFrame conversion for better visualization

print("\033[94m\033[1m"
      "🏗️ Slab Verification (Simplified)\n"
      "------------------------------------------------------------\n"
      f"h (solid) = {hsolid:.3f} cm\n"
      f"I (solid)   = {Isolid:.3f} cm^4\n"
      f"h (waffle) = {htot:.3f} cm\n"
      f"I (waffle)   = {Iwaffle_eq:.3f} cm^4\n"
      f"Iwaffle/Isolid = {Iwaffle_eq/Isolid:.3f}  ->  {'OK ✅' if Slab_ratio >= 1 else 'NOT OK ❌'}\n"
      "\033[0m")
```

#### **3.9. Summary of Results – Beam Preliminary Sizing**

```python
Resul_slab = pd.concat([h_df, bw_df, sr_fd, bf_df, bst_df, hf_df, htot_df, hr_df, AT_df, Iwaffle_eq_df, Isolid_df, Slab_ratio_df, alfam_df], axis=1)  # Combine all slab results into a single DataFrame
Resul_slab.head(len(Resul_slab))                                                                                                    # Display all slab results in a single table
```

## **4.0. Preliminary Sizing of Reinforced Concrete Shear Walls**

---

### Conceptual basis and design inputs (ACI 318-19 framework)

This section presents the **preliminary sizing of reinforced concrete shear walls**, assuming that the **wall geometry in plan (length and location)** is already defined by architectural constraints.  
At this stage, the objective is **not** to design the wall for a specific force demand, but rather to:

- define a **reasonable wall thickness** compatible with:
  - constructability,
  - stiffness requirements,
  - and compatibility with framing beams,
- ensure compliance with **minimum reinforcement requirements** for:
  - flexure (vertical reinforcement),
  - shear and confinement (horizontal reinforcement),
as required by **ACI 318-19**.

This approach is appropriate during early design phases, when global analysis results are not yet finalized, but a structurally sound and code-consistent wall configuration is required.

> **Unit convention adopted in this section**
> - Length: **cm**
> - Forces: **kgf** (internal), **T** (reported)
> - Stresses: **kgf/cm²**

---

#### **4.1. Structural role of shear walls**

Shear walls are vertical structural elements that primarily resist:
- **in-plane shear forces**, and
- **overturning moments** due to lateral loads (seismic and wind).

Unlike beams and slabs, shear walls are **lateral-force-resisting elements**, and their behavior is governed by:
- flexure at the wall base,
- shear along the wall height,
- and axial–flexural interaction.

At the preliminary stage, however, the focus is on ensuring:
- adequate **geometric proportions**, and
- compliance with **minimum reinforcement ratios** required for ductile behavior.

---

#### **4.2. Wall thickness selection criteria**

The **wall thickness $t_w$** is the primary geometric parameter to be selected at this stage.

According to **ACI 318-19, Section 11.7.3**, the minimum thickness of structural walls must satisfy:

$$
t_w \ge \max\left( \frac{l_w}{25},\; 10 \right)
$$

where:
- $t_w$: wall thickness [cm],
- $l_w$: unsupported wall length in plan [cm].

> **Note:** The lower bound of **10 cm** applies to low-demand walls; in practice, structural shear walls in buildings typically adopt larger thicknesses.

---

#### **4.3. Practical thickness limits for building shear walls**

Beyond the code minimum, engineering practice dictates that wall thickness should also consider:

1. **Compatibility with beams framing into the wall**  
   To avoid abrupt stiffness and detailing incompatibilities:
   $$
   t_w \gtrsim b_{\text{beam}}
   $$
   where $b_{\text{beam}}$ is the width of the framing beam.

2. **Constructability and reinforcement congestion**  
   Thin walls may lead to:
   - excessive reinforcement congestion,
   - inadequate concrete placement.

3. **Global stiffness considerations**  
   Walls that are too thin may lead to:
   - excessive lateral drifts,
   - poor dynamic performance.

> **Recommended preliminary range for building shear walls**  
> $$
> 20\ \text{cm} \;\le\; t_w \;\le\; 40\ \text{cm}
> $$

The final value should be **rounded up to a practical construction increment** (e.g., 5 cm).

---

#### **4.4. Minimum vertical reinforcement (flexural reinforcement)**

Vertical reinforcement in shear walls provides:
- flexural strength,
- ductility,
- crack control.

According to **ACI 318-19, Section 11.6.1**, the **minimum vertical reinforcement ratio** for reinforced concrete walls is:

$$
\rho_v \ge 0.0025
$$

Thus, the minimum vertical steel area per unit wall length is:

$$
A_{s,v,\min} = \rho_v \, t_w \, b
$$

where:
- $A_{s,v,\min}$: minimum vertical steel area [cm²],
- $t_w$: wall thickness [cm],
- $b$: wall length considered (typically $b = 100$ cm for a 1 m strip).

> **Design strip convention:**  
> For preliminary checks, reinforcement is evaluated per **1 m wall length**.

---

#### **4.5. Minimum horizontal reinforcement (shear and crack control)**

Horizontal reinforcement in walls contributes to:
- shear resistance,
- crack control,
- confinement of vertical reinforcement.

According to **ACI 318-19, Section 11.6.2**, the minimum horizontal reinforcement ratio is:

$$
\rho_h \ge 0.0015
$$

leading to:

$$
A_{s,h,\min} = \rho_h \, t_w \, b
$$

where:
- $A_{s,h,\min}$: minimum horizontal steel area per meter of wall height [cm²].

---

#### **4.6. Maximum spacing of wall reinforcement**

To ensure effective crack control and confinement, **ACI 318-19, Section 11.7.2**, limits the spacing of wall reinforcement:

For both vertical and horizontal reinforcement:

$$
s \le \min(3t_w,\; 45\ \text{cm})
$$

where:
- $s$: spacing of reinforcement [cm].

This requirement governs bar spacing even when minimum steel ratios are satisfied.

---

#### **4.7. Preliminary reinforcement layout (conceptual)**

At the preliminary stage, a typical wall reinforcement layout consists of:

- **Vertical reinforcement**:
  - uniformly distributed along the wall length,
  - identical reinforcement on both faces of the wall.

- **Horizontal reinforcement**:
  - uniformly distributed along the wall height,
  - identical reinforcement on both faces.

> **Important:**  
> Detailed boundary element requirements, confinement reinforcement, and coupling beam detailing are **not addressed at this stage** and must be evaluated once force demands are available.

#### **Calculations**

```python
varF = 2.00                                                                                                                         # bar diameter for flexural reinforcement in cm
varV = 1.40                                                                                                                         # bar diameter for shear reinforcement in cm
bw = 100.00                                                                                                                         # Analysis Width of wall in cm
Lw = 750.00                                                                                                                         # Length of wall in cm
tw = np.maximum(np.maximum(Lw / 25, 10), b)                                                                                         # Wall thickness in cm

roF = 0.0025                                                                                                                        # Flexural reinforcement ratio
roV = 0.0015                                                                                                                        # Shear reinforcement ratio

AsF = roF * bw * tw                                                                                                                 # Flexural reinforcement area in cm^2 (each face)
AsV = roV * bw * tw                                                                                                                 # Shear reinforcement area in cm^2

Smax = np.minimum( 3 * tw , 45)                                                                                                     # Maximum stirrup spacing in cm

numVarF = AsF / (np.pi * (varF**2) / 4)                                                                                             # Number of flexural bars (each face)
numVarV = AsV / (np.pi * (varV**2) / 4)                                                                                             # Number of shear bars (each face)

numVarF = np.ceil(numVarF)                                                                                                          # Round up to nearest whole number
numVarV = np.ceil(numVarV)                                                                                                          # Round up to nearest whole number

SF = bw / numVarF                                                                                                                   # Spacing of flexural bars in cm
SV = bw / numVarV                                                                                                                   # Spacing of shear bars in cm

SF = np.minimum(SF, Smax)                                                                                                           # Cap flexural bar spacing to max 3*tw or 45 cm
SV = np.minimum(SV, Smax)                                                                                                           # Cap shear bar spacing to max 3*tw or 45 cm

numVarF = np.ceil(bw / SF)                                                                                                          # Recalculate number of flexural bars
numVarV = np.ceil(bw / SV)                                                                                                          # Recalculate number of shear bars

tw_df = pd.DataFrame({'tw (cm)': [tw]})                                                                                             # DataFrame conversion for better visualization
numVarF_df = pd.DataFrame({'Number of Flexural Bars': [numVarF]})                                                                   # DataFrame conversion for better visualization
numVarV_df = pd.DataFrame({'Number of Shear Bars': [numVarV]})                                                                      # DataFrame conversion for better visualization
SF_df = pd.DataFrame({'Flexural Bar Spacing SF (cm)': [SF]})                                                                        # DataFrame conversion for better visualization
SV_df = pd.DataFrame({'Shear Bar Spacing SV (cm)': [SV] })                                                                          # DataFrame conversion for better visualization

varF_df = pd.DataFrame({'Flexural Bar Diameter (mm)': [varF*10]})                                                                   # DataFrame conversion for better visualization
varV_df = pd.DataFrame({'Shear Bar Diameter (mm)': [varV * 10]})                                                                    # DataFrame conversion for better visualization

print("\033[94m\033[1m"
      "🏗️ Wall Summary (Simplified)\n"
      "------------------------------------------------------------\n"
      f"tw = {tw:.3f} cm\n"
      f"# var F   = {numVarF:.3f}\n"
      f"# var V = {numVarV:.3f}"
      "\033[0m")
```

#### **4.8. Summary of Results – Wall Preliminary Sizing**

```python
Resul_wall = pd.concat([tw_df, varF_df, numVarF_df, SF_df, varV_df,numVarV_df, SV_df], axis=1)                                      # Combine all wall results into a single DataFrame
Resul_wall.head(len(Resul_wall))                                                                                                    # Display all wall results in a single table
```
